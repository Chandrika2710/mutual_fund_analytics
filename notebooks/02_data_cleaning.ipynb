```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
nav = pd.read_csv(r"C:\Users\chand\Downloads\mutual_fund_analytics\data\raw\02_nav_history.csv")

performance = pd.read_csv(r"C:\Users\chand\Downloads\mutual_fund_analytics\data\raw\07_scheme_performance.csv")

transactions = pd.read_csv(r"C:\Users\chand\Downloads\mutual_fund_analytics\data\raw\08_investor_transactions.csv")

print("NAV Shape:", nav.shape)
print("Performance Shape:", performance.shape)
print("Transactions Shape:", transactions.shape)
```

```python
# Remove duplicates
nav = nav.drop_duplicates()

# Remove rows with missing NAV values
nav = nav.dropna(subset=["nav"])

# Convert date column
nav["date"] = pd.to_datetime(nav["date"])

# Sort records
nav = nav.sort_values(["amfi_code", "date"])

print(nav.head())
```

```python
nav.to_csv(r"C:\Users\chand\Downloads\mutual_fund_analytics\data\processed\clean_nav.csv", index=False
)

print("clean_nav.csv saved successfully.")
```
